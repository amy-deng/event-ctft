```python
def getRoot(x):
    x = int(x)
    if len(str(x)) == 4: # 1128
        return x // 100
    elif len(str(x)) == 3:
        if x // 10 < 20: # 190
            return x // 10
        else:
            return x // 100
    else:
        return x // 10
    
df['RootEventCode'] = df['CAMEO Code'].apply(lambda x: getRoot(x) )
s_year = 2012
e_year = 2012
lead_window = 1
start = str(s_year) + '-01-10'
end = str(e_year) + '-12-31'

for date_i in pd.date_range(start, end, freq=str(lead_window)+'D'):
    event_date = str(date_i.strftime("%Y-%m-%d"))
    date_j = date_i+pd.Timedelta(days=lead_window) # pd.offsets.Day(2)
    event_date_end = str(date_j.strftime("%Y-%m-%d")) 
    filter_events = df.loc[ (df['Event Date'] >= event_date ) & (df['Event Date'] < event_date_end ) & (df['City']=='Moscow')] 
    
    event_count = filter_events['RootEventCode'].value_counts().to_dict()
#     print(event_count,sum(event_count.values()))
    n_protest = event_count.get(14,0)
    n_all_event = sum(event_count.values())
    print('suseptability: {:.4f} '.format(n_protest/n_all_event))
    
    
```
